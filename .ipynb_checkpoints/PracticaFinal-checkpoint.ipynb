```python
# Importación de librerias
import spacy
import pandas as pd
import pymongo
from pymongo import MongoClient
import certifi
from config import *
```

# Manejo de Datos

```python
datos = pd.read_csv("data\data_total_con_codigo_001.csv") # Extracción de la data
datos
```

```python
datosPrueba = datos.head(100) # Almacenamiento de un cantidad especifica de datos
datosPrueba
```

# Aplicando Spacy

```python
nlp = spacy.load("es_core_news_sm") #Carga de la instancia del modelo de lenguaje
```

## Pre-procesamiento de los datos

```python
documents = [] # Creación de lista

# 1. Elimanación de signos 
for text in datosPrueba['feedback']:
     documents.append(text.lower().replace('\\', '').replace('"','').replace('(','').replace(')','').replace('%','')
                      .replace('$','').replace('>','').replace('\\n','').replace(',','').replace(':','').replace('.','')
                      )

documents[0:5]
```

```python
words = [] # Creación de lista

# 2. Extracción de palabras relevantes
for text in documents:
        myDoc = nlp(text)
        for word in myDoc:
            lexeme = nlp.vocab[word.text]
            if lexeme.is_stop == False:
                words.append(word)
            
uniqueWords = [] # Creación de lista

# 3. Eliminación de palabras repetidas
for w in words:
        if (w.text not in uniqueWords) and (w.like_num is False):
            uniqueWords.append(w.text)

# 4. Eliminación de palabras obsoletas
uniqueWords.remove(' ')
uniqueWords.remove('   ')
uniqueWords.remove('x')
uniqueWords.remove('pbi')
uniqueWords.remove('xy')
uniqueWords.remove('num')
uniqueWords.remove('vs')
uniqueWords.remove('n')
uniqueWords.remove('dce')
uniqueWords.remove('plot')
uniqueWords.remove('ah')
uniqueWords.remove('eh')
uniqueWords.remove('2010-2015')

# 5. Conversión de palabras a objetos de tipo doc
uniqueWords = [nlp(u) for u in uniqueWords]
uniqueWords
```

## Palabras a su forma base

```python
dic = {} #Creación de diccionario

# 6. Aplicando lemmatización y llenar diccionario

for word in uniqueWords:
    for token in word:
        dic[token.text] = token.lemma_
        print(token, ":" ,token.lemma_)
```
